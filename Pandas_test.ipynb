```python
age_mean = {}
age_mean44 = {}
for tf in df2["gender"].unique():
    age_mean1 = df2.loc[df2["gender"]==tf,"age"].mean()
    age_mean2 = df2.loc[df2["gender"]==tf,"age"].count()
    age_mean[tf] = age_mean1
    age_mean44[tf] = age_mean2
```
